Influence of “Category” on “Lifetime Post Consumers. - Anyone doing this?

```python
groupedcat = fb.groupby('Category')
groupedcat = groupedcat.aggregate(np.mean)
sns.barplot(x='Category',y='Lifetime Post Consumers',data=groupedcat.reset_index())
```
